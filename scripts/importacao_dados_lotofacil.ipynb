```python
# Renomear colunas para facilitar a manipulação
df.columns = [col.strip().replace(' ', '_').replace('/', '_').replace('á', 'a').replace('ç', 'c') for col in df.columns]

# Converter a coluna de data
df['Data_Sorteio'] = pd.to_datetime(df['Data_Sorteio'], dayfirst=True, errors='coerce')

# Garantir que as bolas sejam inteiras
for i in range(1, 16):
    df[f'Bola{i}'] = pd.to_numeric(df[f'Bola{i}'], errors='coerce')
```

```python
# Criar a coluna com lista de dezenas
df['Dezenas'] = df[[f'Bola{i}' for i in range(1, 16)]].values.tolist()

# Criar a coluna com string das dezenas separadas por hífen (ex: "01-03-05-...-25")
df['Dezenas_str'] = df['Dezenas'].apply(lambda x: '-'.join(f'{int(i):02d}' for i in sorted(x)))
```

```python
#  1. Frequência Absoluta de Cada Número (1 a 25)
from collections import Counter

# Junta todas as dezenas em uma lista única
todas_dezenas = [dezena for linha in df['Dezenas'] for dezena in linha]

# Conta quantas vezes cada dezena apareceu
frequencia = Counter(todas_dezenas)

# Organiza em DataFrame
df_freq = pd.DataFrame(sorted(frequencia.items()), columns=['Dezena', 'Frequência'])

# Frequência relativa
total_sorteios = len(df)
df_freq['Frequência (%)'] = (df_freq['Frequência'] / total_sorteios) * 100

```

```python
#  2. Gráfico de Barras – Frequência Absoluta
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(data=df_freq, x='Dezena', y='Frequência', palette='viridis')

plt.title('Frequência Absoluta das Dezenas na Lotofácil', fontsize=14)
plt.xlabel('Dezena', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 3. Top 5 – Mais e Menos Sorteadas

mais_sorteadas = df_freq.sort_values('Frequência', ascending=False).head(5)
menos_sorteadas = df_freq.sort_values('Frequência', ascending=True).head(5)

print("🔝 Top 5 Mais Sorteadas:")
print(mais_sorteadas.to_string(index=False))

print("\n🔻 Top 5 Menos Sorteadas:")
print(menos_sorteadas.to_string(index=False))
```

```python
# 4. Gráfico de Barras – Frequência Relativa (%)
plt.figure(figsize=(12,6))
sns.barplot(data=df_freq, x='Dezena', y='Frequência (%)', palette='plasma')

plt.title('Frequência Relativa (%) das Dezenas na Lotofácil', fontsize=14)
plt.xlabel('Dezena', fontsize=12)
plt.ylabel('Frequência (%)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 5. Heatmap das Frequências – Grid 5x5 (simulando volante)
import numpy as np

# Prepara os dados para o grid 5x5
grid_freq = np.zeros((5, 5), dtype=int)

for i, freq in enumerate(df_freq['Frequência']):
    linha = i // 5
    coluna = i % 5
    grid_freq[linha, coluna] = freq

# Cria o heatmap
plt.figure(figsize=(8,6))
sns.heatmap(grid_freq, annot=True, fmt="d", cmap='YlOrRd', 
            xticklabels=[f'Col {i+1}' for i in range(5)],
            yticklabels=[f'Linha {i+1}' for i in range(5)])

plt.title('Heatmap da Frequência das Dezenas no Volante da Lotofácil', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 6. Soma das Dezenas Sorteadas

df['Soma'] = df['Dezenas'].apply(sum)
```

```python
# 7. Quantidade de Pares e Ímpares

df['Qtd_Pares'] = df['Dezenas'].apply(lambda x: sum(1 for n in x if n % 2 == 0))
df['Qtd_Impares'] = df['Dezenas'].apply(lambda x: sum(1 for n in x if n % 2 != 0))
```

```python
# 8. Contagem por Linha e Coluna do Volante

# Funções para contar dezenas por linha
def contar_linhas(dezenas, linha):
    return sum(1 for d in dezenas if (linha - 1)*5 < d <= linha*5)

# Funções para contar dezenas por coluna
def contar_colunas(dezenas, coluna):
    return sum(1 for d in dezenas if (d - 1) % 5 == (coluna - 1))

# Criar colunas com as contagens
for i in range(1, 6):
    df[f'Qtd_Linha_{i}'] = df['Dezenas'].apply(lambda x: contar_linhas(x, i))
    df[f'Qtd_Coluna_{i}'] = df['Dezenas'].apply(lambda x: contar_colunas(x, i))
```

```python
# 9. Número de Repetições em Relação ao Sorteio Anterior

df['Qtd_Repetidos'] = df['Dezenas'].shift(1).combine(df['Dezenas'], 
                            lambda a, b: len(set(a) & set(b)) if isinstance(a, list) else 0)
```

```python
# 10. Análise de somas e padrões pares/ímpares

import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a soma das dezenas sorteadas em cada concurso
df['Soma_Dezenas'] = df['Dezenas'].apply(sum)

# Calcular a quantidade de pares e ímpares
def contar_pares_impares(dezenas):
    pares = sum(1 for d in dezenas if d % 2 == 0)
    impares = 15 - pares
    return pares, impares

df[['Qtd_Pares', 'Qtd_Impares']] = df['Dezenas'].apply(
    lambda x: pd.Series(contar_pares_impares(x))
)

# Estatísticas descritivas das somas
print("📊 Estatísticas descritivas da soma das dezenas:")
print(df['Soma_Dezenas'].describe())

# Histograma da soma das dezenas
plt.figure(figsize=(10, 5))
sns.histplot(df['Soma_Dezenas'], bins=30, kde=True, color='skyblue')
plt.title('Distribuição da Soma das Dezenas')
plt.xlabel('Soma')
plt.ylabel('Frequência')
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico de barras: distribuição pares/ímpares
pares_impares_counts = df[['Qtd_Pares', 'Qtd_Impares']].value_counts().sort_index()

# Reformata para exibir em gráfico
pares_impares_df = pares_impares_counts.reset_index()
pares_impares_df.columns = ['Qtd_Pares', 'Qtd_Impares', 'Frequência']
pares_impares_df['Label'] = pares_impares_df['Qtd_Pares'].astype(str) + 'P-' + pares_impares_df['Qtd_Impares'].astype(str) + 'I'

plt.figure(figsize=(12, 6))
sns.barplot(data=pares_impares_df, x='Label', y='Frequência', palette='viridis')
plt.title('Distribuição de Pares e Ímpares nas Dezenas Sorteadas')
plt.xlabel('Composição Pares-Impares')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#11 Matriz de Coocorrência das Dezenas Sorteadas

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Inicializa a matriz 25x25 com zeros
matriz_cooc = np.zeros((25, 25), dtype=int)

# Percorre cada sorteio e atualiza a matriz
for dezenas in df['Dezenas']:
    for i in dezenas:
        for j in dezenas:
            matriz_cooc[i-1][j-1] += 1  # Subtrai 1 para alinhar ao índice 0-24

# Cria DataFrame para visualização
df_cooc = pd.DataFrame(matriz_cooc, index=range(1, 26), columns=range(1, 26))

# Visualiza com heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(df_cooc, cmap="YlOrRd", linewidths=0.5, square=True)
plt.title('🔁 Matriz de Coocorrência das Dezenas da Lotofácil')
plt.xlabel('Dezena')
plt.ylabel('Dezena')
plt.tight_layout()
plt.show()
```

```python
pares_freq = []

# Percorre a matriz e coleta apenas os pares superiores (sem diagonal nem duplicatas)
for i in range(25):
    for j in range(i+1, 25):
        pares_freq.append(((i+1, j+1), df_cooc.iloc[i, j]))

# Ordena pelos mais frequentes
pares_freq = sorted(pares_freq, key=lambda x: x[1], reverse=True)

# Mostra os 10 principais pares
for par, freq in pares_freq[:10]:
    print(f"Pares {par[0]} e {par[1]} apareceram juntos {freq} vezes")
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Criar DataFrame com estatísticas das dezenas
estatisticas = []

for dezena in range(1, 26):
    freq = df['Dezenas'].apply(lambda x: dezena in x).sum()
    linha = (dezena - 1) // 5 + 1
    coluna = (dezena - 1) % 5 + 1

    estatisticas.append({
        'Dezena': dezena,
        'Frequencia': freq,
        'Linha': linha,
        'Coluna': coluna
    })

df_estat = pd.DataFrame(estatisticas)

# Garantir que os campos são numéricos
df_estat[['Frequencia', 'Linha', 'Coluna']] = df_estat[['Frequencia', 'Linha', 'Coluna']].apply(pd.to_numeric)

# 2. Padronizar os dados
scaler = StandardScaler()
X = scaler.fit_transform(df_estat[['Frequencia', 'Linha', 'Coluna']])

# 3. KMeans clustering com n_init compatível
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_estat['Cluster'] = kmeans.fit_predict(X)

# 4. Plotagem dos clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_estat, x='Coluna', y='Linha', 
                hue='Cluster', palette='Set2', s=100)

# Adicionar os números das dezenas no gráfico
for _, row in df_estat.iterrows():
    plt.text(row['Coluna'] + 0.1, row['Linha'] + 0.1, 
             str(row['Dezena']), fontsize=9)

plt.title('Clusterização das Dezenas da Lotofácil')
plt.xlabel('Coluna')
plt.ylabel('Linha')
plt.grid(True)
plt.legend(title='Cluster')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 12. Análise de Antecessores e Sucessores

from collections import Counter

antecessores = Counter()
sucessores = Counter()

for i in range(1, len(df)):
    dezenas_atuais = set(df.loc[i, 'Dezenas'])
    dezenas_anteriores = set(df.loc[i-1, 'Dezenas'])

    antecessores.update(dezenas_anteriores & dezenas_atuais)
    sucessores.update(dezenas_atuais & set(df.loc[i+1, 'Dezenas']) if i+1 < len(df) else [])

print("Top 10 Antecessores:")
print(antecessores.most_common(10))

print("\nTop 10 Sucessores:")
print(sucessores.most_common(10))
```

```python
# 13. Regra de Associação entre Dezenas

from itertools import combinations
from collections import Counter

contador_pares = Counter()

for dezenas in df['Dezenas']:
    for par in combinations(sorted(dezenas), 2):
        contador_pares[par] += 1

pares_mais_comuns = contador_pares.most_common(10)

print("Top 10 pares de dezenas que mais aparecem juntos:")
for par, freq in pares_mais_comuns:
    print(f"Pares {par[0]} e {par[1]} apareceram juntos {freq} vezes")
```

```python
# 14. Distribuição por década

import pandas as pd

def contar_por_decada(dezenas):
    return {
        '1-10': sum(1 for d in dezenas if 1 <= d <= 10),
        '11-20': sum(1 for d in dezenas if 11 <= d <= 20),
        '21-25': sum(1 for d in dezenas if 21 <= d <= 25),
    }

decadas = df['Dezenas'].apply(contar_por_decada)
df_decadas = pd.DataFrame(list(decadas))

print("Média por década:")
print(df_decadas.mean())
```

```python
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import pandas as pd

# Filtrar últimos 2 anos
df_2anos = df[df['Data_Sorteio'] >= df['Data_Sorteio'].max() - pd.DateOffset(years=2)].copy()

# Extrair colunas de ano e mês
df_2anos['Ano'] = df_2anos['Data_Sorteio'].dt.year
df_2anos['Mês'] = df_2anos['Data_Sorteio'].dt.month
```

```python
freq_mensal = []

for (ano, mes), grupo in df_2anos.groupby(['Ano', 'Mês']):
    dezenas = grupo[[f'Bola{i}' for i in range(1, 16)]].values.flatten()
    contagem = Counter(dezenas)
    for dezena, freq in contagem.items():
        freq_mensal.append({'Ano': ano, 'Mês': mes, 'Dezena': dezena, 'Frequência': freq})

df_freq_mensal = pd.DataFrame(freq_mensal)

# Gráfico
plt.figure(figsize=(16, 6))
pivot = df_freq_mensal.pivot_table(index='Dezena', columns=['Ano', 'Mês'], values='Frequência', fill_value=0)
sns.heatmap(pivot, cmap='YlGnBu', linewidths=0.5)
plt.title('Frequência das Dezenas por Mês (últimos 2 anos)')
plt.xlabel('Ano / Mês')
plt.ylabel('Dezena')
plt.tight_layout()
plt.show()
```

```python
pares_ímpares = df_2anos.groupby(['Ano', 'Mês'])[['Qtd_Pares', 'Qtd_Impares']].mean().reset_index()

# Gráfico
plt.figure(figsize=(12, 5))
sns.lineplot(data=pares_ímpares, x='Mês', y='Qtd_Pares', hue='Ano', marker='o', label='Pares')
sns.lineplot(data=pares_ímpares, x='Mês', y='Qtd_Impares', hue='Ano', marker='s', linestyle='--', label='Ímpares')
plt.title('Média de Pares e Ímpares por Mês (últimos 2 anos)')
plt.xlabel('Mês')
plt.ylabel('Quantidade Média')
plt.tight_layout()
plt.show()
```

```python
soma_mensal = df_2anos.groupby(['Ano', 'Mês'])['Soma_Dezenas'].mean().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(data=soma_mensal, x='Mês', y='Soma_Dezenas', hue='Ano', marker='o')
plt.title('Média da Soma das Dezenas por Mês (últimos 2 anos)')
plt.xlabel('Mês')
plt.ylabel('Média da Soma')
plt.tight_layout()
plt.show()
```

```python
qtd_por_mes = df_2anos.groupby(['Ano', 'Mês']).size().reset_index(name='Qtd_Concursos')

plt.figure(figsize=(12, 5))
sns.barplot(data=qtd_por_mes, x='Mês', y='Qtd_Concursos', hue='Ano')
plt.title('Quantidade de Concursos por Mês (últimos 2 anos)')
plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.tight_layout()
plt.show()
```

```python
top_dezenas = df_freq_mensal.groupby(['Ano', 'Mês']).apply(
    lambda x: x.nlargest(5, 'Frequência')
).reset_index(drop=True)

print("Top 5 dezenas mais frequentes por mês nos últimos 2 anos:")
print(top_dezenas.sort_values(['Ano', 'Mês', 'Frequência'], ascending=[True, True, False]))
```
